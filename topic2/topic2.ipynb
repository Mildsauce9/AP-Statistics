```python
import pandas as pd

data = pd.read_csv('exams.csv')
data.head()
data.describe()
```

```python
import seaborn as sns
g = sns.pairplot(data = data, hue = 'gender', diag_kind = 'kde', palette = 'husl')
g.fig.suptitle('Graph showing distribution between scores and {}'.format('gender'), fontsize=20)
g.fig.subplots_adjust(top=0.9)
```

```python
g = sns.pairplot(data = data, hue = 'race/ethnicity', diag_kind = 'kde', palette = 'husl')
g.fig.suptitle('Graph showing distribution between scores and {}'.format('race/ethnicity'), fontsize=20)
g.fig.subplots_adjust(top=0.9)
```
